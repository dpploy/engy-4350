**Catalog description:** <br>
This course provides an overview of pertinent topics in basic nuclear heat generation and removal in a nuclear      reactor, power conversion, and overall system integration and safety.<br>    
**Prerequisites:** ENGY-4340 Nuclear Reactor Theory, familiarity with thermo, fluids, and heat transfer.<br>
**Co-requisites:** none.<br>
**Course designation:** required. <br>

**Course outcomes:**
 1. Explain principal characteristics of nuclear power reactors such as power cycles, primary coolant systems, components, and advanced reactor design.
 2. Explain and calculate shutdown energy and decay power.
 3. Analyze and calculate thermodynamics of nuclear reactor systems.
 4. Understand thermal properties of nuclear fuels and calculate temperature distribution in a fuel element.
 5. Calculate pressure drop and heat transfer inside a reactor flow channel.
 6. Understand the boiling curve and critical heat flux.
 7. Understand two-phase flow regimes and calculate pressure drop of two-phase flow.
 8. Work in a team to complete a nuclear reactor engineering design project and present the project.
